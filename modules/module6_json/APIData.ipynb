https://www.last.fm/api/show/chart.getTopArtists

```python
import os
from pprint import pprint
from requests import get
from config import API_KEY

url = "http://ws.audioscrobbler.com/2.0/"
params = {
    "method": "chart.gettopartists",
    "api_key": API_KEY, # avoid hard coding secrets
    "format": "json",
    "limit": 10
}
response = get(url, params)

data = response.json()
pprint(data)
```

```python
data['artists']['artist']
```

```python
artists = []

for artist in data['artists']['artist']:
    artists.append(artist["name"])
    
print(len(artists))
print(artists)
```

```python
#http://ws.audioscrobbler.com/2.0/?method=artist.gettopalbums&artist=cher&api_key=YOUR_API_KEY&format=json

url = "http://ws.audioscrobbler.com/2.0/"
params = {
    "method": "artist.gettopalbums",
    "artist": artists[0],
    "api_key": API_KEY,
    "format": "json",
    "limit": 3
}

response = get(url, params)
album_data = response.json()
pprint(album_data)
```

```python
pprint(album_data["topalbums"]["album"])
```

```python
albums = []

for album in album_data["topalbums"]["album"]:
    albums.append(album["name"])
    
print(albums)
```
